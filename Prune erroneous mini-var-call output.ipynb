```python
import os
import fnmatch
import json
import dateutil.parser
import re
import pprint
import glob
import pandas as pd
```

```python
paths = glob.glob("/treehouse/archive/downstream/*/secondary/*mini-var-call*/methods.json")
print("Found {} mini-var-call outputs".format(len(paths)))
```

```python
remove = []
for path in paths:
    m = json.loads(open(path).read())
    if dateutil.parser.parse("1/9/2018") < dateutil.parser.parse(m["start"]) <= dateutil.parser.parse("4/27/2018"):
        remove.append("/pod/pstore/groups" + os.path.split(path)[0])
#         print(m["start"], os.path.split(path)[0])
print("Found {} mini-var-call outputs to remove".format(len(remove)))
```

```python
with open("../mini-var-call-delete.txt", "w+") as f:
    f.write("\n".join(remove))
```
